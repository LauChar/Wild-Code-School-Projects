# Import de la carte et des lieux 

```
import folium
df_lieux = df_lieux[["lieux","Latitude", "Longitude"]]

m = folium.Map(location=[df_lieux.Latitude.mean(), df_lieux.Longitude.mean()], 
                        max_zoom=5, 
                        min_zoom=5, 
                        zoom_start=5, 
                        control_scale=False,
                        zoom_control =False,
                        width = '100%',
                        height = '100%')
```

```
# time line personnages
for lieu in range(8):
    for column in range(1,7):
        if df_places.iloc[lieu,column] != 'NOWHERE':
            idx = dico_lieux['lieux'].index(df_places.iloc[lieu,column])
            folium.Marker(
                location=[dico_lieux["Latitude"][idx], dico_lieux["Longitude"][idx]], 
                popup=df_places.columns[column-1]
            ).add_to(m)
```

```
# définission des lieux
for x in range(18): # attention il faudra mettre le len du dico
    folium.CircleMarker(
        location=[dico_lieux['Latitude'][x],dico_lieux['Longitude'][x]],
        radius=30,
        popup=dico_lieux['lieux'][x],
        color="#3186cc",
        fill=True,
        fill_color="#3186cc",
    ).add_to(m)
```
